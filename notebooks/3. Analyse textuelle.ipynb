## 3. Analyse textuelle  <a class="anchor" id="section_4"></a>

### 3.a Analyse des groupes de 2 mots  <a class="anchor" id="section_4a"></a>

```python
#création d'un dataframe à partir des deux colonnes Commentaire et star 
df_sentiment= df[['Commentaire', 'star']]
#supression des lingnes contenants des valeurs Nan
df_sentiment = df_sentiment.dropna(axis = 0) 
df_sentiment.info()
```

```python
#création colonne sentiment prenant la valeur 1 si le client est satisfait et 0 sinon
df_sentiment['sentiment']=np.where(df_sentiment['star']>2,1,0)
```

```python
#préparation de la variable commentaire avant l'extraction en groupe de mots
df_sentiment['Commentaire']=df_sentiment['Commentaire'].astype(str)# conversion en string
df_sentiment = df_sentiment[df_sentiment['Commentaire'].notnull()]# supression des commentaires vides
# mettre toute les variables en miniscule:
for i in df_sentiment.index:
    df_sentiment['Commentaire'][i] = df_sentiment['Commentaire'][i].lower()
# supression de la ponctuation:   
df_sentiment['Commentaire'] = df_sentiment['Commentaire'].str.replace(r'[^\w\s]+', '',regex=True)
# supprimer les mots vides:
df_sentiment['Commentaire'] = df_sentiment['Commentaire'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words and len(word)>2]))
```

```python
#création de la variable groupe de mot :
import re
from nltk.util import ngrams

df_sentiment['Commentaire'].astype(str)
df_sentiment['groupe de deux mots'] = df_sentiment['Commentaire']

for i in df_sentiment.index:
    s = str(df_sentiment['Commentaire'][i])
    tokens = [token for token in s.split(" ") if token != ""]
    df_sentiment['groupe de deux mots'][i] = list(ngrams(tokens, 2))
```

```python
#séparation du jeu de données en satifait et non satisfait:
df_satisfait = df_sentiment[df_sentiment['sentiment']==1]
df_nonsatisfait = df_sentiment[df_sentiment['sentiment']==0]
```

```python
#création d'une liste de tout les groupe de mot pour chaque jeu de donnnée

Liste_groupe_mots_1 = []

for i in df_satisfait.index :
    for j in range(len(df_satisfait['groupe de deux mots'][i])):
        Liste_groupe_mots_1.append(df_satisfait['groupe de deux mots'][i][j])
```

```python
#création d'un dataframe à partir de la liste de groupe de mots pour les clients satisfaits:
df_Sat_2 = pd.DataFrame({'Liste_groupe_mots': Liste_groupe_mots_1})  
```

```python
df_Sat_2['Liste_groupe_mots'].value_counts()
```

```python
#Affichage des 30 groupes de 2 mots les plus utilisées par les clients satisfaits
from matplotlib import pyplot as plt

plt.figure(figsize=(20,20))

df_Sat_2['Liste_groupe_mots'].value_counts().head(30).plot(kind='bar')
plt.title('Groupes de deux mots les plus utilisés par les clients satisfaits')
plt.show()
```
